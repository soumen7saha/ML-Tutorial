```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    learning_rate=float(input('Enter the learning rate: ')) #0.001
    iterations=int(input('No. of iterations: ')) #1000
    n=len(x)
    #plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(iterations):
        y_predicted=m_curr*x+b_curr
        #plt.plot(x,y_predicted,color='green')
        
        cost_function=(1/n)*sum((y-y_predicted)**2)
        
        m_derivative= -(2/n)*sum(x*(y-y_predicted))
        b_derivative= -(2/n)*sum(y-y_predicted)
        
        m_curr-=learning_rate*m_derivative
        b_curr-=learning_rate*b_derivative
        
        print(f'm={m_curr}, b={b_curr}, cost={cost_function} for iteration {i}')
    return cost_function
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

y1=np.array([3,4,5,1,20])
y_predicted=2*x+3
print(y_predicted)
print(y1-y_predicted)
print(sum((y1-y_predicted)**2))
sum(i**2 for i in (y1-y_predicted))
```

```python
gradient_descent(x,y) #cost decreasing
```

## Find a linear function for given test results using gradient descent

```python
import pandas as pd
from sklearn import linear_model
import math
```

```python
df=pd.read_csv('test_scores.csv')
df.head()
```

```python
x=np.array(df['math'])
x
```

```python
y=np.array(df['cs'])
y
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    rate=float(input('Enter the learning rate: ')) #0.0002
    n=len(x)
    prev_cost=0.0
    for i in range(1000000):
        y_predicted=m_curr*x+b_curr
        cost=(1/n)*sum(y-y_predicted)
        m_curr-=rate*(-(2/n)*sum(x*(y-y_predicted)))
        b_curr-=rate*(-(2/n)*sum(y-y_predicted))
        if math.isclose(cost,prev_cost,rel_tol=1e-20):
            break
        prev_cost=cost
    print(f'm={m_curr}, b={b_curr}, cost={cost}')
```

```python
gradient_descent(x,y)
```
