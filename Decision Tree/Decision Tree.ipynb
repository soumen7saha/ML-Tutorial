```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('salaries.csv')
df
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
df['company_n']=le_company.fit_transform(df['company'])
df['job_n']=le_job.fit_transform(df['job'])
df['degree_n']=le_degree.fit_transform(df['degree'])
df
```

```python
y=df['salary_more_then_100k']
df.drop(['company','job','degree','salary_more_then_100k'],inplace=True,axis='columns')
df # input data frame
```

```python
y # target data frame
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,y,test_size=0.2)
```

```python
model=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```

## Decision Tree Classification on Titanic Dataset

```python
df1=pd.read_csv('titanic.csv')
df1
```

Target Feature: `Survived`

Input Features:
- `Pclass`
- `Sex`
- `Age`
- `Fare`

```python
y=df1['Survived'] #target df
df1.drop(['PassengerId','Survived','Name','SibSp','Parch','Ticket','Cabin','Embarked'],inplace=True,axis='columns')
df1
```

```python
y
```

```python
le_sex=LabelEncoder()
df1['sex_n']=le_sex.fit_transform(df1['Sex'])
df1
```

```python
df1.drop('Sex',inplace=True,axis='columns')
```

```python
df1 #input df
```

```python
df1.isna().sum()
```

```python
df1['Age'].fillna(int(df1['Age'].mean()),inplace=True)
```

```python
df1.isna().sum()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df1,y,test_size=0.2)
```

```python
model=tree.DecisionTreeClassifier(criterion='entropy')
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

## Decision Tree Classification on Candidates Hiring Dataset

```python
df2=pd.read_csv('data1.csv')
df2.head()
```

```python
df2.isna().sum()
```

```python
le_manager=LabelEncoder()
le_passed=LabelEncoder()
```

```python
df2['previouly in Manager position?']=le_manager.fit_transform(df2['Manager at previous position?'])
y=le_passed.fit_transform(df2['Passed hiring process?'])
```

```python
df2.drop(['Manager at previous position?','Passed hiring process?'],axis='columns',inplace=True)
df2
```

```python
y #target df
```

```python
X_train,X_test,y_train,y_test=train_test_split(df2,y,test_size=0.2)
```

```python
model=tree.DecisionTreeClassifier(criterion='entropy')
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
